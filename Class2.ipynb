```python
data = np.log(df/df.shift(1))

data.iloc[0] = 0
data.head(10)
```

```python
beta_SVM = (data[['Market','SVM']].cov()/data['Market'].var()).iloc[0].iloc[1]
beta_SVM
```

```python
beta, alpha = np.polyfit(data['Market'], data['SVM'], 1)
alpha
beta
```

```python
sns.scatterplot(y = 'Market', x = 'SVM', data = data, color = 'BLUE')
plt.show()
```

```python
sns.scatterplot(y = 'SVM', x = 'Market', data = data, label = 'Returns')
sns.lineplot(x = data['Market'], y = alpha + (data['Market']-alpha)*beta, color = 'GREEN', label = 'CAPM Line')
```

```python
rm = data['Market'].mean()*252
rm
cov = data[['Market','SVM']].cov() *252
cov_SVM_market = cov.iloc[0,1]
cov_SVM_market
market_var = data['Market'].var()*252
market_var

SVM_beta_annual = cov_SVM_market / market_var
print('The annualized beta will equal the one calculated at daily returns:',SVM_beta_annual)

rf = 0.025
riskpremium = rm - rf

## CAPM
SVM_capm_return = rf + SVM_beta_annual*riskpremium

print(f"The annualized CAPM return of SVM is {SVM_capm_return*100:.2f}%")
```

```python
sharperatio = (rm-rf)/(data['SVM'].std()*np.sqrt(252))
sharperatio
print(f"Sharpe Ratio: {round(sharperatio,4)}")
```

```python
sns.scatterplot(y = 'Market', x = 'BANX', data = data, color = 'BLUE')
plt.show()
```

```python
sns.scatterplot(y = 'BANX', x = 'Market', data = data, label = 'Returns')
sns.lineplot(x = data['Market'], y = alpha + (data['Market']-alpha)*beta, color = 'GREEN', label = 'CAPM Line')
```

```python
rm = data['Market'].mean()*252
rm
cov = data[['Market','BANX']].cov() *252
cov_BANX_market = cov.iloc[0,1]
cov_BANX_market
market_var = data['Market'].var()*252
market_var

BANX_beta_annual = cov_BANX_market / market_var
print('The annualized beta will equal the one calculated at daily returns:',BANX_beta_annual)

rf = 0.025
riskpremium = rm - rf

## CAPM
BANX_capm_return = rf + BANX_beta_annual*riskpremium

print(f"The annualized CAPM return of BANX is {BANX_capm_return*100:.2f}%")
```

```python
sharperatio = (rm-rf)/(data['BANX'].std()*np.sqrt(252))
sharperatio
print(f"Sharpe Ratio: {round(sharperatio,4)}")
```
